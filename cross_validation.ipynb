```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,KFold,StratifiedKFold,LeaveOneOut,cross_val_score
from sklearn.metrics import accuracy_score
```

```python
df=sns.load_dataset("iris")
```

```python
df.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
```

```python
x=df.drop(["species"],axis =1)
y=df[["species"]]
```

```python
#holdout cross validation
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model1=dt.fit(xtrain,ytrain)
model1.score(xtest,ytest)
```

```python
#prediction
y_pred=model1.predict(xtest)
evaluation_score=accuracy_score(ytest,y_pred)
evaluation_score
```

```python
#kfold cross validation
kfold=KFold(n_splits=6)
```

```python
score=cross_val_score(dt,x,y,cv=kfold)
score
```

```python
score.mean()
```

```python
#StratifiedKFold
StratifiedKFold=StratifiedKFold(n_splits=6)
```

```python
scorE1=cross_val_score(dt,x,y,cv=StratifiedKFold)
scorE1
```

```python
scorE1.mean()
```

```python
#LeaveOneOut
LeaveOneOut=LeaveOneOut()
```

```python
score2=cross_val_score(dt,x,y,cv=LeaveOneOut)
score2
```

```python
score2.mean()
```
